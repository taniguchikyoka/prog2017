さいころを振って４がでたらおわるプログラム．

```python
import random
while True:
 rand_num = random.randint(1,6)
 print(rand_num)
 if rand_num == 4:
    break
 print('.')
```

無限ループを止めるときにbreakでとまる
いきなり書くといきなり止まってしまうので，４がでたら止まるようにしている．
breakがあるとそのあとにあるものは，実行されないでとまるのでピリオドはでてこない．

```python
import random
while True:
 rand_num = random.randint(1,6)
 print(rand_num)
 if rand_num == 4:
    continue
 print('.')
```

continueがあるとそのあとを無視して，最初に戻る．
